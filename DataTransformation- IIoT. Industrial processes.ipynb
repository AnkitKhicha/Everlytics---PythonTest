```python
#See comments in ‘Sample Input’ for column mapping (e.g., ItemId becomes tag__id)

new_df1 = df1.loc[:,['ItemId','ItemName','Comment']].copy()
```

```python
new_df1.head()
```

```python
df2.head()
```

```python
## Dropping the millisecond column.

new_df2 = df2.drop(['Milli Sec'],axis=1)
```

```python
new_df2.head()
```

```python
# Defining the functin to remove unit from comments.

comment = pd.Series(index = new_df1['Comment'])
```

```python
comment.head()
```

```python
# Seperating the Unit.

new_df1['Unit'] = comment.index.str.extract(r'\((\w+)\)',expand=True)
```

```python
new_df1.head()
```

```python
## Adding the data and time as time_stamp column and dropping the original date and time columns in df2.

new_df2.insert(2,'time_stamp',new_df2['Date'].astype(str) + " "+ new_df2['Time'],True)
```

```python
new_df2.drop(columns=['Date','Time'],axis=1,inplace=True)
```

```python
new_df2.head()
```

```python
new_df2.shape
```

```python
new_df1.shape
```

```python
## Using the max function to aggregate the data capture from 5 millisecond to 10 sec as to returnt the hightest value.

new_df2 = new_df2.groupby(['time_stamp'],as_index=False).max()
```

```python
new_df2.head(5)
```

```python
Sample_Output = pd.DataFrame()
d = {}

d['event_timestamp'] = []
d["tag__id"] = []
d["tag__name"] = []
d["tag__desc"] = []
d["tag__value"] = []
d["tag__unit"] = []

a2 = []
for k in d:
    a2.append(k)

alias = ["ItemId", 'ItemId', 'ItemName', 'Comment', 'ItemId','Unit']
for i in range(7):
    for j in range(6):
        d[a2[j]].append(new_df1[alias[j]][i])
    d[a2[0]][i] = new_df2["time_stamp"][i*10]
    d[a2[4]][i] = -212340981234091283470112342
    for j in range(i*10, min(i*10+10, 64)):
        d['tag__value'][i] = max(new_df2[new_df1[alias[2]][i]][j], d['tag__value'][i])

Output = pd.DataFrame(d)
Output
```

```python
Output.to_csv('SampleOutput.csv',index=False,na_rep ='')
```
